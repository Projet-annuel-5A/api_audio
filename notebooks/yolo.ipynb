```python
import torch
import pandas as pd
from ultralytics import YOLO
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f'Using device: {device} -> {torch.cuda.get_device_name(0)}')
```

TRAINING AND VALIDATION

```python
project_dir = '../data'
data_dir = '../data/images'
project_name = 'images_classifier/train'

epochs = 500
imgsize = 48
batchsize = 160
```

```python
model = YOLO('yolov8x-cls.pt')
model.to(device)
```

```python
model.train(data=data_dir, epochs=epochs, imgsz=imgsize, batch=batchsize, device=device, project=project_dir, name=project_name, plots=True)
```

```python
df = pd.read_csv('../data/images_classifier/train/results.csv')
df
```

```python
df.columns = df.columns.str.strip()

plt.figure(figsize=(12, 6))
plt.plot(df['epoch'], df['train/loss'], label='Training Loss')
plt.plot(df['epoch'], df['val/loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(df['epoch'], df['metrics/accuracy_top1'], label='Accuracy Top 1')
plt.plot(df['epoch'], df['metrics/accuracy_top5'], label='Accuracy Top 5')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Top 1 and Top 5 Accuracy')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(df['epoch'], df['lr/pg0'], label='Learning Rate pg0')
plt.plot(df['epoch'], df['lr/pg1'], label='Learning Rate pg1')
plt.plot(df['epoch'], df['lr/pg2'], label='Learning Rate pg2')
plt.xlabel('Epoch')
plt.ylabel('Learning Rate')
plt.title('Learning Rate per Parameter Group')
plt.legend()
plt.grid(True)
plt.show()
```

```python
project_name_m = 'images_classifier_medium/train'

model_m = YOLO('yolov8m-cls.pt')
model_m.to(device)
```

```python
model_m.train(data=data_dir, epochs=epochs, imgsz=imgsize, batch=batchsize, device=device, project=project_dir, name=project_name_m, plots=True)
```
